# Import Data Set

```python
import pandas as pd

#import main data
players_stats = pd.read_csv("players_stats_by_season_full_details.csv")
players_stats
```

```python
# import supporting data

NBA_salary = pd.read_csv("NBA_season1718_salary.csv")
NBA_salary
```

# Data Cleaning

```python
#Check whether the data has empty value.
for i in players_stats.isnull().T.all():
    if i:
        print(i)
        
for i in NBA_salary.isnull().T.all():
    if i:
        print(i)
```

```python
# Combine two data set
NBA_salary_zip = zip(NBA_salary["Player"],NBA_salary["season17_18"])
NBA_salary_dict = dict(NBA_salary_zip)
NBA_salary_dict
```

```python
NBA_salary_dict["Nicolas Batum"]
```

```python
# Set salary column equal to zero, then import data to main data set
players_stats["salary"] = 0

for index, row in players_stats.iterrows():
    if row["Player"] in NBA_salary_dict.keys():
        players_stats["salary"][index] = NBA_salary_dict[row["Player"]]
players_stats
```

# Data Summarizing

```python
# Compute shooting percentage (FGM/FGA)
hit_rate = []

for i,j in zip(players_stats['FGM'],players_stats['FGA']):
    if j != 0:
        hit_rate.append(round(i/j,5))
    else:
        hit_rate.append(0)
        
# Add a new column called 'hit_rate'
players_stats["hit_rate"] = hit_rate  
players_stats
```

```python
# Compute average height, weight and shooting percentage from each basket league
# mean_height_cm、mean_hitRate
League = players_stats.League.unique()

height_cm_mean = []
hit_rate_mean = []
weight_kg_mean = []

for i in League:
    t = players_stats.League == i
    height_cm_mean.append(round(players_stats[t].height_cm.mean(),3))
    hit_rate_mean.append(round(players_stats[t].hit_rate.mean(),3))
    weight_kg_mean.append(round(players_stats[t].weight_kg.mean(),3))

# Average height
height_cm_mean = dict(zip(League,height_cm_mean))
# Average weight
weight_kg_mean = dict(zip(League,weight_kg_mean))
# Average hit rate
hit_rate_mean = dict(zip(League,hit_rate_mean))
```

```python
# Height mean ranks
height_cm_mean_df = pd.DataFrame.from_dict(height_cm_mean, orient='index')
height_cm_mean_df_sorted = height_cm_mean_df.sort_values(by=0 , ascending=False)
height_cm_mean_df_sorted
```

```python
# Weight mean ranks
weight_kg_mean_df = pd.DataFrame.from_dict(weight_kg_mean, orient='index')
weight_kg_mean_df_sorted = weight_kg_mean_df.sort_values(by=0 , ascending=False)
weight_kg_mean_df_sorted
```

```python
# Shooting percentage ranks
hit_rate_mean_df = pd.DataFrame.from_dict(hit_rate_mean, orient='index')
hit_rate_mean_df_sorted = hit_rate_mean_df.sort_values(by=0 , ascending=False)
hit_rate_mean_df_sorted
```
